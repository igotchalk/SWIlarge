```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt 

lithdir = Path('../data/lith')
sgemsdir = Path('../data/lith/sgems')
```

```python
def sgems2modflow(a):
    return np.transpose(a.reshape((345,300,20),order='F'),axes=(2,1,0))
```

```python
fname = sgemsdir.joinpath('MPS','reals','mps180_16node.csv')
a_txt = np.loadtxt(fname,delimiter=',',skiprows=1,dtype=np.str)
a = np.array([1 if s.endswith('1') else 0 for s in a_txt])


np.save(fname.parent.joinpath(fname.name[:-4]), sgems2modflow(a))
```

### Plot

```python
plt.imshow(sgems2modflow(a)[0,:,:])
plt.gca().set_aspect(1)
```
