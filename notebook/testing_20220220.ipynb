```python
from pathlib import Path
import pickle

def get_it_from_fpath(f):
    return f.name.split('_')[1]

def load_obj(dirname,name):
    with open(Path(dirname).joinpath(name + '.pkl').as_posix(), 'rb') as f:
        return pickle.load(f)
```

```python
f_varlist =Path('/home/users/ianpg/git_cloned/SWIlarge/data/PriorModel/varlist.pkl')
ws_outputaem = Path('/scratch/users/ianpg/SWIlarge/work/output/AEM/data/')
jobnum = '45634935'
jobnum2 = '45657535'



if len(f_varlist.name.split('.'))>1:
    nam = f_varlist.name.split('.')[0]
else:
    nam = f_varlist.name
varlist = load_obj(f_varlist.parent,nam)
k = list(varlist.keys())
```

```python
f_varlist2 =Path('/home/users/ianpg/git_cloned/SWIlarge/data/PriorModel/varlist_fromgithub.pkl')
varlist2 = load_obj(f_varlist2.parent,nam)
k = list(varlist2.keys())
```

```python
f_aem,it_aem = list(zip(*[(p,int(get_it_from_fpath(p))) for p in ws_outputaem.iterdir() if jobnum in p.name]))
f_aem2,it_aem2 = list(zip(*[(p,int(get_it_from_fpath(p))) for p in ws_outputaem.iterdir() if jobnum2 in p.name]))

it_aem = list(it_aem)
it_aem.extend(it_aem2)
```
